We are building content base recommender system

```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credit = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
credit.head(1)['cast'].values
```

```python
# Merge both data frame to reduce complixity
movies = movies.merge(credit,on='title')
```

```python
movies.head(3)
```

```python
# important col are genres,id,keywords,title,overview,cast,crew
#We will we keeping the above col
movies = movies[['genres','id','keywords','title','overview','cast','crew']]
```

```python
movies.head()
```

```python
movies.isna().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies['genres'][0]
```

```python
#As the value in the dic is in string format so we need  to convert it into normal dic. To do that we have 
#import a module that does it 
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    genres_list=[]
    for i in ast.literal_eval(obj):
        genres_list.append(i['name'])
    return genres_list
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(4)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(4)
```

```python
ast.literal_eval(movies['cast'][0])
```

```python
def convert_cast(obj):
    cast_list=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            cast_list.append(i['name'])
            counter+=1
        else:
            break
    return cast_list
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
ast.literal_eval(movies['crew'][0])
```

```python
def fetch_Director(obj):
    Director_name=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            Director_name.append(i['name'])
            break
    return Director_name
```

```python
fetch_Director(movies['crew'][0])
```

```python
movies['crew']=movies['crew'].apply(fetch_Director)
```

```python
movies.head(4)
```

```python
#converting overview col into a list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#this function remove the spaces from the words i.e Science Fiction  --> ScienceFiction
def remove_space(obj):
    general_list=[]
    for i in obj:
        general_list.append(i.replace(" ",""))
    return general_list
```

```python
movies['genres']=movies['genres'].apply(remove_space)
movies['keywords']=movies['keywords'].apply(remove_space)
movies['cast']=movies['cast'].apply(remove_space)
movies['crew']=movies['crew'].apply(remove_space)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df= movies[['id','title','tags']]
```

```python
new_df.head()
```

```python
new_df.rename(columns={"id":"movie_id"},inplace=True)
```

```python
new_df['tags'][0]
```

```python
#Combine all the items in tags col
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

## Vectorizering the tags

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
#We have to do streaming i.e convert dancing ---> dance
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem(new_df['tags'][0])
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

Calculate the distance between the vector using cosine distance
if distance is more similarity is less and vice versa

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
#to find the index from new_df and the find the same movie from vectore
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0] 
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
